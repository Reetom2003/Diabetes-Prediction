```python
#To find all the zero values of glucose column
zero_glucose_values = diabetes_dataset[diabetes_dataset['Glucose'] == 0]
zero_glucose_values
```

```python
#To find all the zero values of Blood Pressure column
zero_BloodPressure_values = diabetes_dataset[diabetes_dataset['BloodPressure'] == 0]
zero_BloodPressure_values
```

```python
#To find all the zero values of BMI column
zero_BMI_values = diabetes_dataset[diabetes_dataset['BMI'] == 0]
zero_BMI_values
```

```python
#To find all the zero values of SkinThickness column
zero_SkinThickness_values = diabetes_dataset[diabetes_dataset['SkinThickness'] == 0]
zero_SkinThickness_values
```

```python
#To find all the zero values of Insulin column
zero_Insulin_values = diabetes_dataset[diabetes_dataset['Insulin'] == 0]
zero_Insulin_values
```

```python
# Replacing 0 values with the mean/median of the respective feature
#Glucose
diabetes_dataset["Glucose"] = diabetes_dataset["Glucose"].replace(0,diabetes_dataset["Glucose"].median())
#Blood_Pressure
diabetes_dataset["BloodPressure"] = diabetes_dataset["BloodPressure"].replace(0,diabetes_dataset["BloodPressure"].median())
#BMI
diabetes_dataset["BMI"] = diabetes_dataset["BMI"].replace(0,diabetes_dataset["BMI"].mean())
#Skin_Thickness
diabetes_dataset["SkinThickness"] = diabetes_dataset["SkinThickness"].replace(0,diabetes_dataset["SkinThickness"].mean())
#Insulin
diabetes_dataset["Insulin"] = diabetes_dataset["Insulin"].replace(0,diabetes_dataset["Insulin"].mean())
diabetes_dataset
```

```python
# Splitting the dependent and independent variable
X = diabetes_dataset.drop(["Outcome"],axis = 1)
y = diabetes_dataset["Outcome"]
X
```

```python
y
```

```python
#Splitting the dataset into training and testing dataset
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
training_accuracy = []
test_accuracy = []
for n_neighbors in range(1,11):
    knn = KNeighborsClassifier(n_neighbors = n_neighbors)
    knn.fit(X_train, y_train)

    #check accuracy score
    training_accuracy.append(knn.score(X_train, y_train))
    test_accuracy.append(knn.score(X_test, y_test))
```

```python
plt.plot(range(1,11), training_accuracy, label = "Training_accuracy")
plt.plot(range(1,11), test_accuracy, label = "Test_accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
knn = KNeighborsClassifier(n_neighbors = 9)
knn.fit(X_train, y_train)
print("Training Accuracy:", knn.score(X_train, y_train))
print("Training Accuracy in Percentage:", knn.score(X_train, y_train)*100,"%")
print("Test Accuracy:", knn.score(X_test, y_test))
print("Test Accuracy in Percentage:", knn.score(X_test, y_test)*100,"%")
```

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 0, max_depth=6)
dt.fit(X_train, y_train)
print("Training Accuracy:", dt.score(X_train, y_train))
print("Training Accuracy:", dt.score(X_train, y_train)*100,"%")
print("Test Accuracy:", dt.score(X_test, y_test))
print("Test Accuracy:", dt.score(X_test, y_test)*100,"%")
```

**Support Vector Machine(SVM)**

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear', random_state=0)
svm.fit(X_train, y_train)
print("Training Accuracy:", svm.score(X_train, y_train))
print("Training Accuracy:", svm.score(X_train, y_train) * 100, "%")
print("Test Accuracy:", svm.score(X_test, y_test))
print("Test Accuracy:", svm.score(X_test, y_test) * 100, "%")
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
logistic_reg = LogisticRegression(random_state=0)
logistic_reg.fit(X_train, y_train)
print("Training Accuracy:", logistic_reg.score(X_train, y_train))
print("Training Accuracy:", logistic_reg.score(X_train, y_train) * 100, "%")
print("Test Accuracy:", logistic_reg.score(X_test, y_test))
print("Test Accuracy:", logistic_reg.score(X_test, y_test) * 100, "%")
```

**Multilayer Perceptron(MLP)**

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(random_state=42)
mlp.fit(X_train, y_train)
print("Training Accuracy:", mlp.score(X_train, y_train))
print("Training Accuracy:", mlp.score(X_train, y_train)*100,"%")
print("Test Accuracy:", mlp.score(X_test, y_test))
print("Test Accuracy:", mlp.score(X_test, y_test)*100,"%")
```
